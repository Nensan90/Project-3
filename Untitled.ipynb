```python
# Import Dependencies
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

# X and Y values
x = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
y1 = [1.501, 1.422, 1.352, 1.555, 1.851, 2.281, 2.576, 2.811, 3.263, 2.365, 2.793, 3.528, 3.61, 3.511, 3.376, 2.423]
y2 = [72, 163, 263, 111, 19, 77, 21, 43, 28, 3, 15, 19, 13, 3, 2, 0]
```

```python
# plotting figures by creating axes object
# using subplots() function
fig, ax = plt.subplots(figsize = (10, 5))
plt.title('Avg Gas Price/Gallon vs Unhealthy # Air Quality Days per Year in New York')
```

```python
# using the twinx() for creating another
# axes object for secondary y-Axis
ax2 = ax.twinx()
ax.plot(x, y1, color = 'g')
ax2.plot(x, y2, color = 'b')
```

```python
# giving labels to the axises
ax.set_xlabel('Year', color = 'r')
ax.set_ylabel('Avg Price per Gallon/Year', color = 'g')
```

```python
# secondary y-axis label
ax2.set_ylabel('Total # of Unhealthy Air Quality Days/Year', color = 'b')
```

```python
plt.tight_layout()
plt.show()
```
